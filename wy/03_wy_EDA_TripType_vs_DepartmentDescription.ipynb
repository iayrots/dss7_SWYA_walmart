### Triptype 별 뜯어보기 (39, 40)

```python
# 40 (식재료 + personal care + 안경)
triptype_each = train[train['TripType'].isin([40])]
plt.figure(figsize=(20, 40))
sns.countplot(y='DepartmentDescription', data=triptype_each)
plt.show()
```

```python
# 39 (식재료 + personal care)
triptype_each = train[train['TripType'].isin([39])]
plt.figure(figsize=(20, 40))
sns.countplot(y='DepartmentDescription', data=triptype_each)
plt.show()
```
